## Определение

Функция — это фрагмент программного кода, к которому можно обратиться из другого места программы

Функция представляет некую абстрактность в этом мире, т.е. человек идет, машина едет, птица летит. У каждого объекта, есть определенные фукции с которыми он работает

## Синтаксис и нюансы
Объявление функции в python выглядит следующим оразом 

```python
# объявление функции my_function()
def my_function(arg_1, arg_2):
  # тело функции
```

Где **def** это оператор после которого идет название функции, после названия идут параметры функции, далее на следующей строке через табуляцию(либо 4 пробелла) тело функции.

Небольшое уточнение, параметры часто путают с аргументами:
*  Параметр – это переменная, которой будет присваиваться входящее в функцию значение.
*  Аргумент – само это значение, которое передается в функцию при её вызове.

```python
# a, b - параметры функции
def test (a, b):
    # do something

# 120, 404 — аргументы
test(120, 404)
```



**Ключевая особенность функций** – возможность возвращать значение.

Для этого используется слово **return**. Предположим, вы часто умножаете числа. Вы не осведомлены заранее, целые они или вещественные, но хотите, чтобы результат был целым всегда. Решим задачу с помощью функции:


```python
# она будет принимать два множителя, а возвращать их округленное 
# до целого числа произведение
def int_multiple(a, b):
    product = a * b
    # возвращаем значение
    return int(product)

print(int_multiple(341, 2.7))
> 920
```



````python
def int_multiple(a, b):
    product = a * b
    return int(product)

print(int_multiple(341, 2.7))
````
